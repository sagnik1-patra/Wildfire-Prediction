```python
import os
import json
import yaml
import joblib
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout


# ======================================================
# ðŸ“‚ Paths
# ======================================================
DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Wildfire Prediction\archive\forestfires.csv"
SAVE_DIR = r"C:\Users\NXTWAVE\Downloads\Wildfire Prediction"
VISUAL_DIR = os.path.join(SAVE_DIR, "visuals")

os.makedirs(SAVE_DIR, exist_ok=True)
os.makedirs(VISUAL_DIR, exist_ok=True)

# ======================================================
# ðŸ“Œ Load Dataset
# ======================================================
print("[INFO] Loading dataset...")

df = pd.read_csv(DATA_PATH)

lbl_month = LabelEncoder()
lbl_day = LabelEncoder()

df["month"] = lbl_month.fit_transform(df["month"])
df["day"] = lbl_day.fit_transform(df["day"])

X = df.drop("area", axis=1)
y = df["area"]

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, os.path.join(SAVE_DIR, "hybrid_firelens_scaler.pkl"))

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)


# ======================================================
# ðŸ§  Build DL Model
# ======================================================
def build_model(params):
    model = Sequential([
        Dense(params["units1"], activation="relu", input_shape=(X_train.shape[1],)),
        Dropout(params["drop1"]),
        Dense(params["units2"], activation="relu"),
        Dropout(params["drop2"]),
        Dense(1)
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=params["lr"]),
        loss="mse"
    )
    return model


# ======================================================
# ðŸ’‰ AIS (Artificial Immune System) Operator
# ======================================================
def ais_clone_and_mutate(params):
    new_params = params.copy()

    # Random mutation
    if random.random() < 0.5:
        new_params["units1"] = random.randint(32, 128)
    if random.random() < 0.5:
        new_params["units2"] = random.randint(16, 64)
    if random.random() < 0.5:
        new_params["drop1"] = random.uniform(0.1, 0.4)
    if random.random() < 0.5:
        new_params["drop2"] = random.uniform(0.1, 0.4)
    if random.random() < 0.5:
        new_params["lr"] = random.uniform(0.0005, 0.01)
    if random.random() < 0.5:
        new_params["batch"] = random.choice([16, 32, 64])

    return new_params


# ======================================================
# ðŸ¥š CSA (Cuckoo Search Algorithm)
# ======================================================
def csa_new_solution(params):
    new_params = params.copy()

    # Levy-flight inspired random jumps
    jump_scale = random.uniform(0.5, 1.5)

    new_params["units1"] = int(params["units1"] * jump_scale)
    new_params["units2"] = int(params["units2"] * jump_scale)

    new_params["drop1"] = min(0.4, max(0.1, params["drop1"] * jump_scale))
    new_params["drop2"] = min(0.4, max(0.1, params["drop2"] * jump_scale))

    new_params["lr"] *= random.uniform(0.8, 1.2)

    return new_params


# ======================================================
# ðŸ§ª Fitness Function
# ======================================================
def fitness(params):
    model = build_model(params)
    model.fit(X_train, y_train, epochs=20, batch_size=params["batch"], verbose=0)
    pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    return rmse


# ======================================================
# ðŸ”¥ Hybrid AIS + CSA Optimization
# ======================================================
def random_params():
    return {
        "units1": random.randint(32, 128),
        "units2": random.randint(16, 64),
        "drop1": random.uniform(0.1, 0.4),
        "drop2": random.uniform(0.1, 0.4),
        "lr": random.uniform(0.0005, 0.01),
        "batch": random.choice([16, 32, 64])
    }


POP = 10
GEN = 6

population = [random_params() for _ in range(POP)]
best_params = None
best_score = float("inf")

print("\n[INFO] Running Hybrid AIS + CSA Optimizer...")

for gen in range(GEN):
    print(f"\n[GENERATION {gen+1}] ---------------------")

    new_population = []

    for params in population:
        score = fitness(params)

        # Track best
        if score < best_score:
            best_score = score
            best_params = params

        # AIS clone & mutate
        clone = ais_clone_and_mutate(params)

        # CSA jump
        cuckoo = csa_new_solution(params)

        new_population.extend([params, clone, cuckoo])

    # Select best POP solutions
    new_population_sorted = sorted(
        new_population,
        key=lambda p: fitness(p)
    )

    population = new_population_sorted[:POP]

print("\n[INFO] BEST PARAMETERS FOUND:")
print(best_params)


# ======================================================
# ðŸ”¥ Train Final Model
# ======================================================
final_model = build_model(best_params)
history = final_model.fit(
    X_train, y_train,
    epochs=50, batch_size=best_params["batch"], verbose=1
)

final_model.save(os.path.join(SAVE_DIR, "hybrid_firelens_model.h5"))


# ======================================================
# ðŸ“¢ Predictions
# ======================================================
y_pred = final_model.predict(X_test).reshape(-1)

result_df = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})

result_df.to_csv(os.path.join(SAVE_DIR, "hybrid_firelens_result.csv"), index=False)

with open(os.path.join(SAVE_DIR, "hybrid_firelens_prediction.json"), "w") as f:
    json.dump({"data": result_df.to_dict()}, f, indent=4)

with open(os.path.join(SAVE_DIR, "hybrid_firelens_config.yaml"), "w") as f:
    yaml.dump(best_params, f)


# ======================================================
# ðŸ“Š VISUALS
# ======================================================

# 1ï¸âƒ£ Heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Hybrid AIS + CSA â€” Heatmap")
plt.savefig(os.path.join(VISUAL_DIR, "hybrid_firelens_heatmap.png"))
plt.close()

# 2ï¸âƒ£ Loss Graph
plt.figure(figsize=(8, 5))
plt.plot(history.history["loss"])
plt.title("Hybrid AIS + CSA â€” Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.savefig(os.path.join(VISUAL_DIR, "hybrid_firelens_loss.png"))
plt.close()

# 3ï¸âƒ£ Actual vs Predicted
plt.figure(figsize=(8, 5))
plt.plot(result_df["Actual"], label="Actual", linewidth=2)
plt.plot(result_df["Predicted"], label="Predicted", linewidth=2)
plt.legend()
plt.title("Hybrid AIS + CSA â€” Actual vs Predicted")
plt.savefig(os.path.join(VISUAL_DIR, "hybrid_firelens_comparison.png"))
plt.close()

# 4ï¸âƒ£ Result Bar Graph
plt.figure(figsize=(12, 5))
result_df[:50].plot(kind="bar")
plt.title("Hybrid AIS + CSA â€” First 50 Results")
plt.savefig(os.path.join(VISUAL_DIR, "hybrid_firelens_results.png"))
plt.close()

# 5ï¸âƒ£ Prediction Scatter
plt.figure(figsize=(8, 5))
plt.scatter(result_df["Actual"], result_df["Predicted"], alpha=0.6)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Hybrid AIS + CSA â€” Prediction Scatter Plot")
plt.savefig(os.path.join(VISUAL_DIR, "hybrid_firelens_prediction_graph.png"))
plt.close()

print("\nðŸŽ‰ ALL FILES SAVED WITH hybrid_ PREFIX!")
print("Location:", SAVE_DIR)
```
